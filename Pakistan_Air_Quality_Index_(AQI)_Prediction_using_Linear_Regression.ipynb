## **Pollutant Components (Input Features)**
These features measure the concentration of harmful gases and particles in the atmosphere.

**components_pm2_5 (Fine Particulate Matter)**: Particles smaller than $2.5 \mu m$ that can enter the respiratory system. It is the leading cause of smog in Pakistan.

**components_pm10 (Coarse Particulate Matter)**: Particles smaller than $10 \mu m$, including dust, pollen, and larger smoke particles.

**components_no2 (Nitrogen Dioxide)**: A harmful gas emitted mainly from traffic exhaust and industrial power plants.

**components_so2 (Sulfur Dioxide)**: A gas produced by burning fossil fuels (coal and oil) and industrial manufacturing.

**components_co (Carbon Monoxide)**: A colorless, odorless gas formed by incomplete fuel combustion in vehicles.

**components_o3 (Ozone)**: Ground-level ozone created by chemical reactions between other pollutants and sunlight.

**components_nh3 (Ammonia)**: A gas primarily released from agricultural fertilizers and organic waste.



##**Meteorological Features**
**temperature_2m (Temperature)**: Air temperature measured 2 meters above ground, affecting the chemical reaction rates of pollutants.


**relative_humidity_2m (Humidity)**: The moisture level in the air. High humidity often traps pollutants near the surface during the winter smog season.


**wind_speed_10m (Wind Speed)**: The speed of air at 10 meters height. Higher speeds help in the dispersion and dilution of pollutants.


**shortwave_radiation (Solar Radiation)**: The intensity of sunlight, which acts as a catalyst for the formation of ground-level Ozone.

##4. **Data Loading and Initial Inspection**

````python
import pandas as pd
import numpy as np

# Load the combined Pakistan dataset
train_path = '/content/aqi_data/Training/concatenated_dataset_Aug_2021_to_July_2024.csv'
df = pd.read_csv(train_path)

print("✅ Data Loaded: Combined Pakistan Dataset")
print("\n--- Columns Available ---")
print(df.columns.tolist())

print("\n--- Data Head ---")
print(df.head())
````

````python
# Run this to see all column names
print(df.columns.tolist())
````

##5. **Data Preprocessing**

````python
# 5.1 Handling Missing Values
df.fillna(df.mean(numeric_only=True), inplace=True)

# 5.2 Feature Selection (Using your exact column names)
features = [
    'components_pm2_5', 'components_pm10', 'components_no2',
    'components_so2', 'components_co', 'components_o3',
    'temperature_2m', 'relative_humidity_2m'
]
target_col = 'main_aqi'

X = df[features]
y = df[target_col]

# 5.3 Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("✅ Features and Target selected successfully!")
````

##6. **Exploratory Data Analysis (EDA)**

````python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
sns.heatmap(df[features + [target_col]].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation: Pollutants vs Main AQI (Pakistan)")
plt.show()
````

##**Model Building – Linear Regression**

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Split data
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train Model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# 9. Coefficients Analysis
coeff_df = pd.DataFrame({'Feature': features, 'Impact': lr_model.coef_})
print("\n--- Pollutant Impact Analysis ---")
print(coeff_df.sort_values(by='Impact', ascending=False))
````

##**Model Evaluation**

````python
# Prediction
y_pred = lr_model.predict(X_val)

# Metrics
mae = mean_absolute_error(y_val, y_pred)
r2 = r2_score(y_val, y_pred)

print(f"\n--- Model Performance ---")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"R² Score (Accuracy): {r2:.2f}")

# 11.2 Actual vs Predicted Plot
plt.figure(figsize=(8,6))
plt.scatter(y_val, y_pred, alpha=0.4, color='darkblue')
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'r--', lw=2)
plt.xlabel("Actual AQI")
plt.ylabel("Predicted AQI")
plt.title("Actual vs Predicted AQI Values")
plt.show()
````

##**Residual Analysis**

````python
residuals = y_val - y_pred
plt.figure(figsize=(10,5))

# Residual Distribution
plt.subplot(1, 2, 1)
sns.histplot(residuals, kde=True, color='red')
plt.title("Error Distribution")

# Homoscedasticity
plt.subplot(1, 2, 2)
plt.scatter(y_pred, residuals, alpha=0.3)
plt.axhline(0, color='black', linestyle='--')
plt.title("Residuals vs Predictions")

plt.tight_layout()
plt.show()
````
